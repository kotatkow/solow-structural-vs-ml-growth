```python
import sys
import os

sys.path.append(os.path.abspath(".."))

import numpy as np
import matplotlib.pyplot as plt
from src.simulation.solow import solow_simulation, steady_state

alpha=0.33; s=0.2; n=0.02; g=0.02; delta=0.05
T=100; dt=0.1

k_path = solow_simulation(alpha=alpha, s=s, n=n, g=g, delta=delta, T=T, dt=dt)
k_star = steady_state(alpha=alpha, s=s, n=n, g=g, delta=delta)

y_path = k_path**alpha
y_star = k_star**alpha

t = np.arange(len(k_path)) * dt

plt.figure()
plt.plot(t, k_path, label="k(t)")
plt.axhline(k_star, linestyle="--", label="k*")
plt.title("Capital per effective worker converges to steady state")
plt.xlabel("time")
plt.ylabel("k")
plt.legend()
plt.show()

plt.figure()
plt.plot(t, y_path, label="y(t)=k(t)^alpha")
plt.axhline(y_star, linestyle="--", label="y*")
plt.title("Output per effective worker converges to steady state")
plt.xlabel("time")
plt.ylabel("y")
plt.legend()
plt.show()

alpha=0.33; s=0.2; n=0.02; g=0.02; delta=0.05

k_star = steady_state(alpha=alpha, s=s, n=n, g=g, delta=delta)

k_grid = np.linspace(1e-6, max(5*k_star, 1.0), 500)
investment = s * (k_grid**alpha)                 # s f(k)
breakeven = (n + g + delta) * k_grid             # (n+g+δ)k

plt.figure()
plt.plot(k_grid, investment, label="s f(k)")
plt.plot(k_grid, breakeven, label="(n+g+δ) k")
plt.axvline(k_star, linestyle="--", label="k*")
plt.title("Break-even investment and steady state")
plt.xlabel("k")
plt.ylabel("per-period investment")
plt.legend()
plt.show()

alpha=0.33; n=0.02; g=0.02; delta=0.05
T=100; dt=0.1
s_list = [0.1, 0.2, 0.3]

plt.figure()
for s in s_list:
    k_path = solow_simulation(alpha=alpha, s=s, n=n, g=g, delta=delta, T=T, dt=dt)
    plt.plot(np.arange(len(k_path))*dt, k_path, label=f"s={s}")
plt.title("Convergence paths for different savings rates")
plt.xlabel("time")
plt.ylabel("k")
plt.legend()
plt.show()

alpha=0.33; s=0.2; g=0.02; delta=0.05
T=100; dt=0.1
n_list = [0.01, 0.02, 0.04]

plt.figure()
for n in n_list:
    k_path = solow_simulation(alpha=alpha, s=s, n=n, g=g, delta=delta, T=T, dt=dt)
    plt.plot(np.arange(len(k_path))*dt, k_path, label=f"n={n}")
plt.title("Convergence paths for different population growth rates")
plt.xlabel("time")
plt.ylabel("k")
plt.legend()
plt.show()

s=0.2; n=0.02; g=0.02; delta=0.05
T=100; dt=0.1
alpha_list = [0.25, 0.33, 0.40]

plt.figure()
for alpha in alpha_list:
    k_path = solow_simulation(alpha=alpha, s=s, n=n, g=g, delta=delta, T=T, dt=dt)
    plt.plot(np.arange(len(k_path))*dt, k_path, label=f"alpha={alpha}")
plt.title("Convergence paths for different capital shares")
plt.xlabel("time")
plt.ylabel("k")
plt.legend()
plt.show()
```
